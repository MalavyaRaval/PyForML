### Task 1.1: Comprehensions (no unnecessary loops)

Create (without loops):
- `even_squares`: squares of even cnumbers from 1..20
- `pairs`: list of (x, x*x) for x=1..10

**Explain:** Why can comprehensions be faster than manual loops?

```python
# TODO:

'''
Comprehensions are faster than manual loops this is because it runs on C internally and it does not need to keep appending and can create a list directly.

'''
# Create even_squares
# HINT: range(1, 21), filter even with x % 2 == 0
even_squares = [x**2 for x in range(1, 21) if x%2 == 0] # Time complexity O(n) n = 20

# TODO:
# Create pairs
pairs = [(x, x*x) for x in range(1, 11)] # Time complexity O(n) n = 10
print(pairs)

# TODO:
# Flatten nested
nested = [[1, 2], [3, 4], [5, 6]]
# HINT: nested loops inside a single comprehension
flattened = [x for y in nested for x in y] # Time complexity O(nm), n = 3 m = 2

check('even_squares', even_squares == [4, 16, 36, 64, 100, 144, 196, 256, 324, 400])
check('pairs', pairs[:3] == [(1, 1), (2, 4), (3, 9)] and pairs[-1] == (10, 100))
check('flattened', flattened == [1, 2, 3, 4, 5, 6])
```

### Task 2.1: Two Sum (Hash Map Pattern)

**FAANG Classic.** Return indices of two numbers that sum to target.

# HINT:
- Use a dict: value -> index
- For each `x`, check if `target-x` has been seen

**Explain:** Why is this O(n) time?

```python
'''

Time Complexity: O(n)
Space Complexity: O(n)

O(n) time because we are goining over the input nums once and the operations of adding values in hashmap and fetching values from hashmap are O(1), so the time complexity is O(1)
'''
def two_sum(nums, target):
    # TODO
    hm = {}

    for i, n in enumerate(nums): # Time complexity O(n) n = len(nums)
        hm[n] = i

    for i in range(len(nums)): # Time complexity O(n) n = len(nums)
        if target - nums[i] in hm:
            if hm[target - nums[i]] != i: # Time complexity O(1) to search in hm
                return [i, hm[target - nums[i]]]

check('two_sum_1', two_sum([2, 7, 11, 15], 9) == [0, 1])
check('two_sum_2', two_sum([3, 2, 4], 6) == [1, 2])
check('two_sum_dup', two_sum([3, 3], 6) == [0, 1])
```

### Task 2.2: Longest Substring Without Repeating Characters (Sliding Window)

**FAANG Classic/Gotcha:** naive O(n²) solutions time out.

# HINT:
- Keep a map char -> most recent index
- Maintain a window `[l..r]` with unique chars

**Explain:** What invariant does the window maintain?

```python
'''

Time Complexity: O(n)
Space Complexity: O(1)

The window always has unique characters in it... It maintains it by checking if the new character that wants to come in is already persent in the hashset/the window or not, if it is present then we keep removeing the elements from the hashset/window from the other side.
'''

def length_of_longest_substring(s: str) -> int:

    hs = set()

    slow = 0
    fast = 0

    res = 0

    while fast < len(s): # Time complexity O(n) n = len(s)
        if s[fast] not in hs:
            hs.add(s[fast])
            fast += 1
            res = max(res, fast - slow)

        else:
            while slow < fast and s[fast] in hs: # Time complexity O(n) n=len(s)
                hs.remove(s[slow])
                slow += 1

    return res
    # TODO


check('ls_abcabcbb', length_of_longest_substring('abcabcbb') == 3)
check('ls_bbbbb', length_of_longest_substring('bbbbb') == 1)
check('ls_pwwkew', length_of_longest_substring('pwwkew') == 3)
check('ls_empty', length_of_longest_substring('') == 0)
```

### Task 2.3: Product of Array Except Self (Prefix/Suffix)

**FAANG Classic/Gotcha:**
- No division
- O(n) time

# HINT:
- Build prefix products in one pass
- Multiply by suffix products in reverse pass

**Explain:** Why does this work with zeros?

```python
'''

Time Complexity: O(n)
Space Complexity: O(1)

Only the position with zero in the original array gets the product of all other elements, positions with non-zero numbers get 0 because multiplication includes the zero
'''

def product_except_self(nums):
    # TODO
    res = [1]*len(nums)
    prod = 1

    for i in range(len(nums)): # Time complexity O(n) n = len(nums)
        res[i] = prod
        prod = prod*nums[i]

    prod = 1
    for i in range(len(nums)-1, -1, -1): # Time complexity O(n) n = len(nums)
        res[i] = prod*res[i]
        prod = prod*nums[i]

    return res


check('pes_1234', product_except_self([1, 2, 3, 4]) == [24, 12, 8, 6])
check('pes_zero', product_except_self([-1, 1, 0, -3, 3]) == [0, 0, 9, 0, 0])
```

### Task 3.1: Sliding Window Maximum (Deque)

**FAANG Classic:** maintain a decreasing deque of indices.

# HINT:
- Pop from left if out of window
- Pop from right while current value is larger

**Explain:** Why is this amortized O(n)?

```python
'''

Time Complexity: O(n)
Space Complexity: O(k)

Since we are going over the input nums we are only adding it the the queue only once so O(n) time and we keep removing the element if the previous element is not bigger than the new element, this way we keep the maximum.

The time complexity in deque to remove from both sides is O(1) so addiding and removing elements is O(1). and hence the overall time complexity is O(n)+O(1)
and Space Complexity = O(k)
'''

def max_sliding_window(nums, k):
    # TODO
    q = deque()
    res = []

    for i, n in enumerate(nums): # Time complexity O(n) n = len(nums)

        if q and q[0] < i - k + 1:
            q.popleft()

        while q and nums[q[-1]] < n:
            q.pop()

        q.append(i)

        if i >= k-1:
            res.append(nums[q[0]])
    return res

check('msw', max_sliding_window([1,3,-1,-3,5,3,6,7], 3) == [3,3,5,5,6,7])
```
